```python
import pandas as pd
from sklearn.datasets import load_iris

iris = load_iris()

df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['species'] = iris.target

species_map = {0: 'setosa', 1: 'versicolor', 2: 'virginica'}
df['species_name'] = df['species'].map(species_map)

print("=" * 55)
print("IRIS DATASET EXPLORATION")
print("=" * 55)

print("\n[ 1 ] First 5 rows of the dataset:")
print("-" * 55)
print(df.head())

print("\n[ 2 ] Dataset Shape:")
print("-" * 55)
rows, cols = df.shape
print(f"  Rows    : {rows}")
print(f"  Columns : {cols}")
print(f"  (Total {rows} samples, {cols} columns)")

print("\n[ 3 ] Basic Statistical Information:")
print("-" * 55)
print(df.describe())

print("\n[ 4 ] Dataset Info (data types & non-null counts):")
print("-" * 55)
print(df.info())

print("\n[ 5 ] Species Distribution:")
print("-" * 55)
print(df['species_name'].value_counts())

print("\n[ 6 ] Missing Values Check:")
print("-" * 55)
print(df.isnull().sum())

print("\n" + "=" * 55)
print("EXPLORATION COMPLETE")
print("=" * 55)
```
